```python
# Import pandas for data manipulation
import pandas as pd

# Import matplotlib for plotting graphs
import matplotlib.pyplot as plt

# Import seaborn for advanced data visualization
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Display the first 5 rows
df.head()
```

```python
# Display information about the dataset
df.info()
```

```python
# Generate statistical summary of numerical columns
df.describe()
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Remove columns that are not useful for prediction
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)

# Display first five rows after removing columns
df.head()
```

```python
# Convert categorical values into numerical values
df["diagnosis"] = df["diagnosis"].map({
    "M": 1,
    "B": 0
})

# Display first five rows
df.head()
```

```python
# Plot the number of malignant and benign samples
sns.countplot(x="diagnosis", data=df)

# Add chart title
plt.title("Diagnosis Count")

# Display graph
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(12,10))

# Plot correlation matrix
sns.heatmap(df.corr(), cmap="coolwarm")

# Display graph
plt.show()
```

```python
# Save the cleaned dataset into a new CSV file
df.to_csv("clean_data.csv", index=False)

print("Clean dataset saved successfully.")
```
